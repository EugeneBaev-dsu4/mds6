Задание.

При фиксированном seed=42 поразбирайтесь и поэкспериментируйте с параметрами алгоритма дифференциальной эволюции: strategy, popsize, tol, mutation, recombination и updating.

Постройте графики количества итераций оптимизации функции ackley от значения параметра.

Дополнительное задание.
Найти множество параметров (strategy=..., ..., updating=...), которые требуют минимального количества итераций.

```python
import numpy as np
from scipy.optimize import differential_evolution
from matplotlib import pyplot as plt
```

### Варьирование параметра popsize

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

bounds = [(-10, 10), (-10, 10)]

popsize_list = list(range(5,40))
popsize_list_use = list(map(lambda x: differential_evolution(ackley, bounds, seed=42, popsize=x).nit, popsize_list))
plt.plot(popsize_list,popsize_list_use)
plt.show()
```

### Варьирование параметра tol

```python
tol_list = np.arange(0.01, 0.2, 0.01)
tol_list_use = list(map(lambda x: differential_evolution(ackley, bounds, seed=42, tol=x).nit, tol_list))
plt.plot(tol_list,tol_list_use)
plt.show()
```

### Варьирование параметра recombination

```python
recombination_list = np.arange(0.1,0.9,0.01)
recombination_list_use = list(map(lambda x: differential_evolution(ackley, bounds, seed=42, recombination=x).nit, recombination_list))
plt.plot(recombination_list,recombination_list_use)
plt.show()
```

### Варьирование параметра mutation

```python
mutation_list = np.arange(0.2,0.8,0.01)
mutation_list_use = list(map(lambda x: differential_evolution(ackley, bounds, seed=42, mutation=x).nit, mutation_list))
plt.plot(mutation_list,mutation_list_use)
plt.show()
```

### Варьирование параметра strategy

```python
strategy_list = ['best1bin','best1exp','rand1exp','randtobest1exp','currenttobest1exp','best2exp','rand2exp','randtobest1bin','currenttobest1bin','best2bin','rand2bin','rand1bin']
strategy_list_use = list(map(lambda x: differential_evolution(ackley, bounds, seed=42, strategy=x).nit, strategy_list))
strategy_list_use
```
